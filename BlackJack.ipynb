```python
values = {"Two":2,"Three":3,"Four":4,"Five":5,"Six":6,"Seven":7,"Eight":8,"Nine":9,"Ten":10,"Jack":10,"Queen":10,"King":10,"Ace":11}
suits = {"Heart","Club","Spade","Diamond"}
```

```python
class Cards:
    
    def __init__(self,suit,rank):
        self.suit = suit # heart,club etc
        self.rank = rank # Two, Three etc
        self.value = values[rank] # 2, 3 etc
        
    def __str__(self):
        return self.rank + " Of "+ self.suit
```

```python
from random import shuffle
class Deck:
    
    def __init__(self):
        self.all_cards = []
        for i in suits:
            for k in values.keys():
                self.all_cards.append(Cards(i,k))
        
    
    def shuffle(self):
        shuffle(self.all_cards)
    
    def draw_card(self):
        return self.all_cards.pop(0)
    
    #def add_cards(self,card):
    #    self.all_cards.append(card)
    
    def __str__(self):
        return (f'Deck has {len(self.all_cards)} cards')
```

```python
class Player:
    
    def __init__(self,name,chip_count):
        self.name = name
        self.chip_count = chip_count
    
    def add_chip(self,chips):
           self.chip_count += chips    
    
    def take_chips(self,chips):
         self.chip_count = self.chip_count - chips
    
    def __str__(self):
        return(f'{self.name} has {self.chip_count} chips')
```

```python
# Create a deck
# Ask player name
# Distribute 2 cards faceup to player
# distribute one card face up to computer
# Take user input : Hit/Stay
# if stay, draw one card for computer, else count user score
def calculate_score(cards_list):
    sum = 0 
    ace_found = False
    ace_index = 0
    for index,card in enumerate(cards_list):
        if(card.rank == 'Ace'):
            ace_found = True
            ace_index = index
        else:
            sum = sum + card.value
    
    if(ace_found):
        sum_final = sum + cards_list[ace_index].value
        if(sum_final <=21):
            return sum_final
        else:
            return (sum +1)
    
    return sum
    

print('Welcome to BlackJack !!!')
name = input('Please provide player name: ')

while True:
    try:
        chips_count = int(input('How many chips you want to deposit. Each hand is 10 chips: '))
    except:
        print('Not correct chip value')
    else:
        break
pl = Player(name,chips_count)
while (pl.chip_count >= 10): 
        dk = Deck()
        dk.shuffle()
        print('***************************************************************************')
        print(dk)
        print('Deck shuffled !!')
        print(f'Hello {pl.name} !! You have {pl.chip_count} chips left')
        print('Playing a hand with 10 chips')
        pl_cards_list = []
        cm_cards_list = []
        pl_card1 = dk.draw_card()
        pl_card2 = dk.draw_card()
        pl_cards_list.append(pl_card1)
        pl_cards_list.append(pl_card2)
        print(f'{pl.name} card 1 : {pl_card1}')
        print(f'{pl.name} card 2 : {pl_card2}')
        pl_score = calculate_score(pl_cards_list)
        comp_card1 = dk.draw_card()
        comp_card2 = dk.draw_card()
        cm_cards_list.append(comp_card1)
        cm_score = calculate_score(cm_cards_list)
        pl_busted = False
        cm_busted = False
        print(f'Computer card 1 : {comp_card1}')
        hs = ''
        pl_card3 = dk.draw_card()
        print(f'Player score: {pl_score}')
        print(f'Computer score: {cm_score}')
        while hs not in ['H','S','s','h','Hit','Stay','HIT','STAY']:
            hs = input('Do you want to Hit or Stay (H/S) :')
        if(hs in ['H','Hit','HIT']):
            print(f'{pl.name} card 3 : {pl_card3}')
            pl_cards_list.append(pl_card3)
            pl_score = calculate_score(pl_cards_list)
            if(pl_score > 21):
                print('Player Busted !!!')
                pl_busted = True
        
        print(f'Computer card 2 : {comp_card2}')
        cm_cards_list.append(comp_card2)
        cm_score = calculate_score(cm_cards_list)
        if(cm_score > 21):
            print('Computer Busted !!!')
            cm_busted = True

        if(cm_score > pl_score or pl_busted):
            print(f'Player score: {pl_score}')
            print(f'Computer score: {cm_score}')
            print('Player Lost !!! ')
            pl.take_chips(10)
            continue
        
        while(cm_score < 17):
            new_cm_card = dk.draw_card()
            cm_cards_list.append(new_cm_card)
            print(f'Computer NEW card : {new_cm_card}')
            cm_score = calculate_score(cm_cards_list)
       
        if(cm_score > 21):
            print('Computer Busted !!!')
            cm_busted = True
            
        if(cm_busted or pl_score > cm_score):
            print(f'Player score: {pl_score}')
            print(f'Computer score: {cm_score}')
            print('Player Wins !!!')
            pl.add_chip(10)
            continue 
        else:
            if(pl_score == cm_score):
                print('Match Tie !!!')
                continue
            else:
                print(f'Player score: {pl_score}')
                print(f'Computer score: {cm_score}')
                print('Player Lost !!! ')
                pl.take_chips(10)
                continue
else:
    print('Your chips count is low !!')
    
        
        
        
    
   
    



```
