```python
def display_board(matrix):    
    a = matrix
    print(f'{a[0]} |{a[1]} |{a[2]}')
    print('--------')
    print(f'{a[3]} |{a[4]} |{a[5]}')
    print('--------')
    print(f'{a[6]} |{a[7]} |{a[8]}')

def welcome_TicTakToe():
    print('Welcome to Tic Tak Toe ! \n Please note the cell positions to play !!')
    display_board(['1','2','3','4','5','6','7','8','9'])
    player1 = ''
    player2 = ''
    while player1 not in ['X','O']:    
        player1 = input('Player1 pls provide your sign X or O: ')
    if(player1 == 'X'):
        player2 = 'O'
    else:
        player2 = 'X'
    print(f'Player1 play with {player1} symbol')
    print(f'Player2 play with {player2} symbol')
    return [player1,player2]

def user_input(curr_player,matrix):
    a = curr_player
    us_i = ''
    inRange = False
    while us_i.isdigit() == False or inRange == False:
        isRange = False
        us_i = input(f'{curr_player} please provide a number 1-9: ')
        
        if(us_i.isdigit()):
            if(int(us_i) in range(1,10)):
                if(matrix[int(us_i)-1] == ' '):
                    inRange = True
                else:
                    print('Cell is already occupied !!')
    return int(us_i)

def update_board(symbol,user_input,matrix):
    matrix[user_input-1] = symbol
    from IPython.display import clear_output
    clear_output()
    display_board(matrix)
    return matrix

def game_on_check():
    y_or_n = 'N'
    y_or_n = input('Do you want to quit Y/N:')
    if(y_or_n == 'Y'):
        print('Thanks for playing !!')
        return False
    else:
        print('Continue !!')
        return True
    
def play_TicTakToe(player_choices):    
    game_on = True
    p1 = player_choices['player1']
    p2 = player_choices['player2']
    matrix = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
    while game_on:
        u1 = user_input('player1',matrix)
        matrix = update_board(p1,u1,matrix)
        if(not identify_winner(matrix,p1,'player1') or not(check_space_avl(matrix))):
            break;       
       
        u2 = user_input('player2',matrix)
        matrix = update_board(p2,u2,matrix)
        if(not identify_winner(matrix,p2,'player2') or not(check_space_avl(matrix))):
            break;       
       # game_on = game_on_check()
        if(game_on == False):
            break
    return False
def identify_winner(matrix,symbol,player):
    win_com = symbol*3
    x = [index for index,value in enumerate(matrix) if value == symbol]   
    x = map(str,x)
    placed = ''.join(x)    
    winners = ['012','345','678','036','147','258','048','256']
    for k in winners:
        if k in placed:
            print(f'{player} wins !!!')
            return False
    return True

def check_space_avl(matrix):
    for i in matrix:
        if i== ' ':
            return True
    print('No more space available ! Match Tie !!')
    return False
    
```
